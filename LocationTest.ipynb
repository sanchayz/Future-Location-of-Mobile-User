```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn import metrics
coords=cdata.as_matrix(columns=['lat','long'])
print (coords)
kms_rad=200
epsilon=0.1/kms_rad
print(epsilon)
db=DBSCAN(eps=epsilon, min_samples=10, algorithm="ball_tree",metric="haversine").fit(np.array(coords))
cluster_labels=db.labels_
num_clusters=len(set(cluster_labels)-set([-1]))
```
